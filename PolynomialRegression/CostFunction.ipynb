```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def compute_cost(X, y, theta):
    """
    Computes the cost for logistic regression.

    Parameters:
        X: numpy array of shape (m, n+1), where m is the number of examples, and n is the number of features.
        y: numpy array of shape (m,), the actual labels (0 or 1).
        theta: numpy array of shape (n+1,), the model parameters.

    Returns:
        J: The cost value.
    """
    m = len(y)  # Number of training examples
    h = sigmoid(X.dot(theta))  # Predicted probabilities
    epsilon = 1e-5  # Small constant to avoid log(0)
    J = (-1 / m) * np.sum(y * np.log(h + epsilon) + (1 - y) * np.log(1 - h + epsilon))
    return J
```

```python
# Example data
X = np.array([[1, 2], [1, 3], [1, 4], [1, 5]])  # Feature matrix (with bias term)
y = np.array([0, 0, 1, 1])  # Labels
theta = np.array([0, 0])  # Initial parameters (weights)

# Compute the cost
cost = compute_cost(X, y, theta)
print(f"Initial Cost: {cost:.4f}")
```

```python
# Visualization of cost function
y_values = [0, 1]  # Actual labels
p = np.linspace(0.01, 0.99, 100)  # Predicted probabilities
costs = []

for y in y_values:
    cost = -y * np.log(p) - (1 - y) * np.log(1 - p)
    costs.append(cost)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(p, costs[0], label="y = 0")
plt.plot(p, costs[1], label="y = 1")
plt.title("Logarithmic Loss (Cost Function)")
plt.xlabel("Predicted Probability")
plt.ylabel("Cost")
plt.legend()
plt.grid()
plt.show()
```
