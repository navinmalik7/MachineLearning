```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Actual values (what the fruit really is)
actual = ["Apple", "Orange", "Orange", "Apple", "Orange", "Apple", "Apple"]

# Predicted values (what the model predicted)
predicted = ["Apple", "Orange", "Apple", "Apple", "Orange", "Orange", "Apple"]

# Create the Confusion Matrix
cm = confusion_matrix(actual, predicted, labels=["Apple", "Orange"])

# Visualize the Confusion Matrix
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Apple", "Orange"], yticklabels=["Apple", "Orange"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for Fruit Classification")
plt.show()
```
